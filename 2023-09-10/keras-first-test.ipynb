```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
dataset = loadtxt('./data/diabetes.data.csv', delimiter=',')
```

```python
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# 8:12:8:1 => [[422,78],[102,106]]
# 8:16:32:16:8:1 => [[402,98],[64,204]]
# 8:32:256:32:1 =>  [[438,62],[87,181]]

model = Sequential()
model.add(Dense(512, input_shape=(8,), activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=300, batch_size=20)
```

```python
loss, accuracy = model.evaluate(X, y)
print(loss, accuracy)
```

```python
prediction = model.predict(X)
yp = [0 if e < 0.5 else 1 for e in prediction]
cm = confusion_matrix(y, yp)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1])
disp.plot()
```
