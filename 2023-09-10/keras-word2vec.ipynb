```python
import tensorflow as tf
import numpy as np

embeddings_index = {}
with open('/home/tina/Downloads/glove.6B.100d.txt') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
```

```python
text = "Using pre-trained word embeddings in a TensorFlow model is relatively straightforward. The first step is to download the pre-trained embeddings from a source such as the GloVe website or the Word2Vec website."

def text2vec(text):
    vector = []
    for w in text.lower().split():    
        v = embeddings_index.get(w)
        if v is not None:
            vector.append(v)
    return vector
    
print(text2vec(text))
```
