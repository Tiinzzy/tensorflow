```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
import time

import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
file_name = 'USUnemployment.csv'

month_mapping = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,'May': 5, 'Jun': 6, 
                 'Jul': 7, 'Aug': 8,'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

mdf = pd.read_csv('./data/' + file_name)
```

```python
mdf.head()
```

```python
def reshape_dataframe(df, sort=True):
    melted_df = df.melt(id_vars=['Year'], var_name='Month', value_name='Percentage')
    melted_df['YMD'] = melted_df['Year'].astype(str) + '-' + melted_df.apply(lambda x: str(month_mapping[x.Month]).zfill(2), axis=1) + '-01'
    if sort:
        melted_df = melted_df.sort_values(by='YMD')
    return melted_df
```

```python
df = reshape_dataframe(mdf)
```

```python
df.head(12)
```

```python
def plot_dataframe(df):
    plt.figure(figsize=(30, 6))
    plt.plot(df['YMD'], df['Percentage'])
    plt.xlabel('YMD')
    plt.ylabel('Percentage')
    plt.title('Unemployment Percentage Over Time')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.xticks(np.arange(0, len(df)+1, 20))
    plt.show()

plot_dataframe(df)
```

```python
def add_time_as_x(df):
    df['time_x'] = df.apply(lambda x: time.mktime(dt.datetime.strptime(x.YMD, "%Y-%m-%d").timetuple()), axis=1)
    return df
```

```python
def create_linear_regression_model(df):
    df_copy = df.copy()
    df_copy = add_time_as_x(df_copy)

    model = LinearRegression()
    X = df_copy[['time_x']].values
    y = df_copy.Percentage.values     
    model.fit(X, y)

    predictions = model.predict(X)
    df_copy['Predicted_Percentage'] = predictions

    plt.figure(figsize=(30, 6))
    plt.scatter(df_copy['YMD'], df_copy['Percentage'], label='Actual', s=2)
    plt.plot(df_copy['YMD'], df_copy['Predicted_Percentage'], color='red', label='Predicted')
    plt.xlabel('YMD')
    plt.ylabel('Percentage')
    plt.legend()
    plt.title('Actual vs. Predicted Percentage')
    plt.xticks(rotation=45)
    plt.xticks(np.arange(0, len(y)+1, 20))
    plt.grid(True)

    plt.show()

create_linear_regression_model(df)
```

```python
def creat_head_map(df):
    df_copy = df.copy()
    df_copy.Month = df_copy.apply(lambda x: str(month_mapping[x.Month]).zfill(2), axis=1)
    
    pivot_table = df_copy.pivot_table(index='Month', columns='Year', values='Percentage')
    
    plt.figure(figsize=(32, 5))
    sns.heatmap(pivot_table, annot=False, cmap='YlGnBu', fmt=".1f", cbar_kws={'label': 'Percentage'})
    plt.title('Heatmap of Unemployment Percentage')
    plt.xlabel('Year')
    plt.ylabel('Month')
    plt.show()

creat_head_map(df)
```

```python
def cluster_to_find_simillarities(df):
    df_copy = df.copy()
    pivot_table = df_copy.pivot_table(index='Year', columns='Month', values='Percentage')
    
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(pivot_table)
    
    
    n_clusters = 5  
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(scaled_data)
    
    pivot_table['Cluster'] = clusters
    
    plt.figure(figsize=(30, 6))
    
    palette = sns.color_palette('husl', n_clusters)
    
    for cluster_id in range(n_clusters):
        cluster_data = pivot_table[pivot_table['Cluster'] == cluster_id]
        plt.scatter(cluster_data.index, [cluster_id] * len(cluster_data), label=f'Cluster {cluster_id}', color=palette[cluster_id])
    
    plt.title('Clustering of Years Based on Monthly Data')
    plt.xlabel('Year')
    plt.ylabel('Cluster')
    plt.legend()
    plt.grid(True)
    plt.show()

cluster_to_find_simillarities(df)
```
