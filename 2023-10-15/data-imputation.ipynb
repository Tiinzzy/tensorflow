```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
```

```python
file_name = 'USUnemployment.csv'
df = pd.read_csv('./data/' + file_name)

pd.set_option('display.max_rows', df.shape[0]+1)

def make_zero(df, row_fraction, col_id):
    no_of_rows = len(df) // row_fraction
    rows = np.random.choice(df.index, no_of_rows)
    df.loc[rows, [col_id]] = np.nan    
    return df

df = make_zero(df, 3, 'Sep')
df = make_zero(df, 3, 'Oct')
```

```python
df.head()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

idf = pd.DataFrame(imp.fit_transform(df))
idf.columns = df.columns
idf.index = df.index

idf.head()
```

```python
imp = KNNImputer(n_neighbors=2,weights='uniform')
idf = pd.DataFrame(imp.fit_transform(df))
idf.columns = df.columns
idf.index = df.index

idf.head()
```
